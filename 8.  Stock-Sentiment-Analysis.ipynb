# Stock Sentiment Analysis using News Headlines

```python
import pandas as pd
df=pd.read_csv('Data.csv', encoding = "ISO-8859-1") 
#with the help of this special encoding, we can see this paticular  dataset
df.head()
```

o means stock price will not going to increase.
1 means stock price will going to increase 

```python
train = df[df['Date'] < '20150101']   #taking all the dataset that is below this paticular date(20150101) as train
test = df[df['Date'] > '20141231']    #taking all the dataset that is after this paticular date(20141231) as test
```

```python
# Removing punctuations
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
```

```python
# Renaming column names for ease of access
list1= [i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns= new_Index
data.head(5)
```

```python
# Convertng headlines to lower case
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

##### converting all the coulmns into a single paragraph because text preprocessing is all about converting the sentences into vectors so that we can apply on machiene learning algorithms on this paticular vector itself 

```python
headlines = []
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[0]
```

```python
headlines[1]
```

#### countvectorizer takes the snetences and it converts it into vector of feature. Also it helps in creating bag of words  

### example of a count vectorizer 

this is how they convert sentences into vectors 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

*we can use naive bais or multi-naive bais instead of randomforest*

```python
## implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(headlines)
```

### for ngram range = (2,2) - we will have two coulmns for Y. like [the cat], [the dog]


```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = randomclassifier.predict(test_dataset)
```

```python
print(predictions)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

# solving the same problem using multi-naive bias 

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive=MultinomialNB()
naive.fit(traindataset,train['Label'])
```

```python
## Predict for the Test Dataset
test_transform= []
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset = countvector.transform(test_transform)
predictions = naive.predict(test_dataset)
```

```python
print(predictions)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
